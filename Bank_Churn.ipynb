```python
# Random UnderSampling
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
# Using the random_state to get the Same Number
random_under_sampling = RandomUnderSampler(random_state=2529)
```

```python
# Using the Under Sampling Method:
X_random_under, Y_random_under = random_under_sampling.fit_resample(X,y)
```

```python
X_random_under.shape, Y_random_under.shape, X.shape, y.shape
```

```python
y.value_counts()
```

```python
class_0 = Y_random_under[Y_random_under == 0]
class_1 = Y_random_under[Y_random_under == 1]

# Determine the Range of the Bins:
bin_range = np.array([-0.25, 0.25, 0.75, 1.25])

# Plotting the Histogram:
plt.hist([class_0, class_1], bins=bin_range, label=['Class 0', 'Class 1'], align='mid')
plt.legend()
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.title('Histogram of Classes')
plt.xticks([0,0.25,0.5,0.75,1])
plt.show()
```

```python
# Using the Train Test Split Method:
```

```python
from sklearn.model_selection import train_test_split
```

```python
# When we split the Original DataSets:
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=2529)
```

```python
# Standardize the Features
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Instantiating the class StandardScaler():
sc = StandardScaler()
```

```python
# Standardize the Original Data:
```

```python
# Using the X_train DataSets and Transform them Accordingly:
X_train[['CreditScore', 'Age', 'Tenure', 'Balance', 'Estimated Salary']]=sc.fit_transform(X_train[['CreditScore','Age','Tenure','Balance','Estimated Salary']])
```

```python
# Using the X_test DataSets and Transform them Accordingly:
X_test[['CreditScore', 'Age', 'Tenure', 'Balance', 'Estimated Salary']]=sc.fit_transform(X_test[['CreditScore','Age','Tenure','Balance','Estimated Salary']])
```

```python
# Support the Vector Machine Classifier:
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
# Training the Machine Learning Model
svc.fit(X_train,y_train)
```

```python
# Predicting the DataSets using the Machine Learning Model that is being Created:
y_pred = svc.predict(X_test)
```

```python
# Checking for the Model Accuracy:
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter_grid = {'C':[0.1,1,10],
                  'gamma':[1,0.1,0.01],
                  'kernel':['rbf'],
                  'class_weight':['balanced']}
```

```python
grid = GridSearchCV(SVC(), parameter_grid, refit=True,verbose=2, cv=2)
grid.fit(X_train,y_train)
```

```python
# Using the Prediction for the X_test Here using the Trained Machine Learning Model
grid_predictions = grid.predict(X_test)
```

```python
confusion_matrix(y_test,grid_predictions)
```

```python
print(classification_report(y_test, grid_predictions))
```

```python
# Explanation of the Code
# The Model correctly classified the 2159 Instances as True Negative
# The Model Incorrectly classified the 343 Instances as Negative when they are Actually Positive.
# The Model correctly classified 243 Instances as True Positive.
# The Model Incorrectly classified the 255 as Positive when they are Actually Negative.

# For class 0, there is 86% of the Instances predicted as class 0 were actually class 0
# For class 1, there is 49% of the Instances predicted as class 1 were actually class 1
# In Practical Terms, this also mean that the Model is able to Perform Well in Identifying the
# Instances of class 0 but Perform Poorly in Identifying the Instances for class 1.
# Out of all the Instances that were predicted for the class 1.

# Note that the f1-score is also Much Higher in the class 0 as compared to class 1.
# Precision = True Positive /(True Positive + False Positive)
# High Precision mean that the Model tend to make Positive Predictions Correctly, Minimizing the False Positives.
```
