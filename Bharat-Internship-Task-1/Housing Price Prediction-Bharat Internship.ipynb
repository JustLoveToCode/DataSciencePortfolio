```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df.columns
```

```python
# Creating the x Variable
x = df.drop('price', axis=1)
```

```python
x.shape
```

```python
# Creating the y Variable:
y= df["price"]
```

```python
y.shape
```

```python
# Splitting the Data into the Training and the Testing Sets:
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=50)
```

```python
# Checking the Shape of the x_train
x_train.shape
```

```python
# Checking the Shape of the y_train:
y_train.shape
```

```python
# Checking the Shape of the x_test:
x_test.shape
```

```python
# Checking the Shape of the y_test:
y_test.shape
```

```python
# Creating the Linear Regression Model:
model = LinearRegression()

# Training the Linear Regression Model using the train data:
model.fit(x_train, y_train)
```

```python
# Using the Trained Model to predict on the test data:
predictions = model.predict(x_test)
```

```python
predictions.shape
```

```python
x_test = x_test.values.ravel()
y_test = y_test.values.ravel()
predictions = predictions.ravel()
```

```python
MSE = mean_squared_error = (y_test, predictions)
```

```python
r2 = r2_score(y_test, predictions)
```

```python
r2
```

```python
plt.scatter(y_test, predictions)
plt.xlabel("Actual Housing Prices")
plt.ylabel("Predicted Housing Prices")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='green', linestyle='-', linewidth=2, label='Best Fit Line')
plt.title("Housing Actual vs.Predicted Prices")
plt.show()
```
